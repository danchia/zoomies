```python
import numpy as np
import scipy
from scipy.cluster.vq import kmeans2
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
raw_map = []
with open(r"c:\projects\zoomies\data\map\map.txt") as f:
  for l in f.readlines()[1:]:
    fl = [float(x) for x in l.split(' ')]
    assert len(fl) == 2
    raw_map.append(fl)

raw_map = np.array(raw_map)
```

```python
plt.figure()
plt.scatter(raw_map[:, 0], raw_map[:, 1])
plt.show()
```

```python
centroid, label = kmeans2(raw_map, 25, iter=50, minit='++')

plt.figure()
plt.scatter(centroid[:, 0], centroid[:, 1])
plt.show()
```
